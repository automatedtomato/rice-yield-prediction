```python
import xarray as xr
import pandas as pd

era5_dir = '../data/raw/era5/'
ds_narita = xr.open_dataset(era5_dir+'era5_narita.nc')
print(ds_narita)
```

```python
climate_df = pd.concat([climate_narita, climate_asahi, climate_ichihara, climate_katori, climate_sanmu])
climate_df.head()
```

```python
climate_df.drop(columns=['number', 'expver'], inplace=True)
climate_df.head()
```

```python
climate_df.rename(columns={
    't2m': 'temp_2m',
    'stl1': 'soil_temp_l1',
    'swvl1': 'soil_water_vol_l1',
    'ssr': 'net_solar_radiation',
    'tp': 'total_rain'},
    inplace=True)
climate_df.head()
```

```python
climate_df.to_csv(era5_dir + 'concat.csv')
```

```python
import requests
import json
APP_ID='3c65e7ad0ccdf88a2a13a709427c33c4d2fe8c89'
API_VERSION='3.0'
SURVEY_NAME='作物統計調査'
STATS_CODE='00500215'
```

```python
endpoint = f"https://api.e-stat.go.jp/rest/{API_VERSION}/app/json/getStatsList"
params = {
        "appId": APP_ID,
        "statsCode": STATS_CODE,
        "surveyYears": '2023',
        "searchWord": "水稲 市町村別 千葉県", # タイトル検索用キーワード
        "limit": 10
    }
print(f"\n--- 2023年の統計表ID検索 ---")
print(f"URL: {endpoint}")
print(f'Params: {params}')
```

```python
response = requests.get(endpoint, params=params)
data = response.json()

result = data.get("GET_STATS_LIST", {}).get("RESULT", {})
status = result.get("STATUS", -1)
print(f"API Status (getStatsList): {status}")
```

```python
stats_data_id = None
    # 正常終了の場合のみIDを抽出
if status == 0:
    datalist_inf = data.get("GET_STATS_LIST", {}).get("DATALIST_INF", {})
    table_inf_list = datalist_inf.get("TABLE_INF", [])
    if isinstance(table_inf_list, dict): table_inf_list = [table_inf_list] # 1件の場合

    if table_inf_list:
        # 簡単な選択ロジック (最初の候補を選択)
        # 必要に応じてタイトル等で絞り込むロジックを追加してください
        selected_table = table_inf_list[0]
        stats_data_id = selected_table.get('@id')
        title = selected_table.get('TITLE', {}).get('$') or selected_table.get('TITLE', 'N/A')
        print(f"候補選択: ID={stats_data_id}, Title={title}")
    else:
            print("該当する統計表が見つかりませんでした。")
elif status == 1:
        print("該当する統計表が見つかりませんでした (結果0件)。")
else:
        print(f"APIエラー発生 (getStatsList): {result.get('ERROR_MSG')}")
```

```python
endpoint = f"https://api.e-stat.go.jp/rest/{API_VERSION}/app/json/getStatsData"
params = {
    "appId": APP_ID,
    "statsDataId": stats_data_id,
    "cdCat01": '1010,1013,1017,1034,1035', # 市町村コード (このDBではcat01)
    "cdCat02": '1006',   # 水稲収穫量kg/haコード (このDBではcat02)
    "lang": "J",
    "metaGetFlg": "Y",          # メタ情報も取得
    "cntGetFlg": "N",
    "annotationGetFlg": "Y"
}
print(f"\n--- 統計データ取得 (ID: {stats_data_id}) ---")
print(f"URL: {endpoint}")
print(f"Params: {params}") # 必要なら表示
```

```python

# APIリクエスト実行
response = requests.get(endpoint, params=params)
data = response.json() # JSONとして解釈

# 結果のステータスコード取得
result = data.get("GET_STATS_DATA", {}).get("RESULT", {})
status = result.get("STATUS", -1)
print(f"API Status (getStatsData): {status}")
```

```python
data_inf = None
metadata_inf = None
# 正常終了の場合のみデータを抽出
if status == 0:
    print("データ取得成功")
    statistical_data = data.get("GET_STATS_DATA", {}).get("STATISTICAL_DATA", {})
    data_inf = statistical_data.get("DATA_INF", {}).get("VALUE", [])
    metadata_inf = statistical_data.get("CLASS_INF", {}) # メタ情報
    if isinstance(data_inf, dict): data_inf = [data_inf] # 1件の場合
    total_num = statistical_data.get("RESULT_INF", {}).get("TOTAL_NUMBER", 0)
    print(f"データ件数: {total_num}")
elif status == 1:
        print("データが見つかりませんでした。")
else:
        print(f"APIエラー発生 (getStatsData): {result.get('ERROR_MSG')}")
```
